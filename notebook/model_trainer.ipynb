### Check for multicolinearity

```python
def corr(df , thresh):
    c_c = set()
    data = df.corr()
    for i in range(len(df.columns)):
        for j in range(i):
            if abs(data.iloc[i,j]) > thresh:
                col_name= data.columns[i]
                c_c.add(col_name)
    return c_c
```

### Train-Test-Splilt

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y , test_size=.30 , random_state=51)
```

### Using Standard Scalar for Standarization

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
def evalute_model(true,predicted):
    acc = accuracy_score(true,predicted)
    mat = confusion_matrix(true,predicted)
#    rmse = np.sqrt(mean_squared_error(true,predicted))
    rep = classification_report(true,predicted)
    return acc,mat,rep
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
models = {
    'LogisticRegression' : LogisticRegression(),
    'DecisionTreeClassifier':DecisionTreeClassifier(),
    'SVC':SVC(),
    'RandomForestClassifier':RandomForestClassifier(),
    'GaussianNB':GaussianNB(),
    'KNeighborsClassifier':KNeighborsClassifier(n_neighbors=4)
}
```

```python

trained_model_list=[]
model_list=[]
accuracy_list=[]
n=1

for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(X_train_scaled,y_train)
    

    y_pred=model.predict(X_test_scaled)

    acc,mat,rep=evalute_model(y_test,y_pred)

    print(list(models.keys())[i])
    model_list.append(list(models.keys())[i])

    print('Model Training Performance')
    print("accuracy_score:",acc*100)
    print("confusion_matrix:",mat)
    print("classification_report",rep)

    accuracy_list.append(acc)
    
    print('='*35)
    print('\n')
    
```

```python
a=accuracy_list
b=model_list
```

```python
result = pd.DataFrame(b,columns=['Model'])
result['acccuracy'] = a
result
```

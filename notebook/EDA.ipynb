## *Check for multicollinearity*
### Function For finding feature having correlation higher than threshold

```python
def corr(df1 , threshold):
    c_c = set() #sets used to avoid updation of duplicate feature
    data = df1.corr()
    for i in range(len(data.columns)):
        for j in range(i):
            if abs(data.iloc[i,j]) > threshold:
                col_name = data.columns[i]
                c_c.add(col_name)
    return c_c
```

```python
corr(df1=df,threshold=.85)
```

```python
corr(df,.90)
```

```python
corr(df,.95)
```
